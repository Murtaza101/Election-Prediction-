```python
locations=[]
for x in range(len(df['user'])):
    locations.append(df.user[x]['location'])

locations[0:10]
```

```python
new_locations = []

for location in locations:
    if location.lower() == 'new delhi' or location.lower() == 'new delhi, india':
        new_locations.append('new delhi')
    else:
        if len(location) != 0:
            x = location.split()[0]
            new_locations.append(x.lower())
        else:
            new_locations.append('')
    
new_locations
```

```python
lat, lon = [],[]

for city in new_locations:
    for x in range(len(list_of_cities)):
        if city == list_of_cities['city'][x].lower():
            lat.append(list_of_cities['lat'][x])
            lon.append(list_of_cities['lat'][x])
        else:
            lat.append('')
            lon.append('')
```
